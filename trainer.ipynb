```python
import numpy as np 
import pandas as pd 

'''

place, place_geo_location, place_popularity, keywords
'''

places = pd.read_csv('output.csv')
```

```python
places['location'] = places['location'].apply(lambda x: x.split())
places['des'] = places['des'].apply(lambda x:x.split())
```

```python

places
```

```python
places['location'] = places['location'].apply(lambda x:[i.replace(',','') for i in x])
places['location'] = places['location'].apply(lambda x:[i.replace(' ','') for i in x])
places['des'] = places['des'].apply(lambda x:[i.replace(',','') for i in x])
places['des'] = places['des'].apply(lambda x:[i.replace(' ','') for i in x])
places
```

```python
places['tags'] = places['location'] + places['des']
places
```

```python
new_df = places[['name','cords','tags']]
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x))
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text): 
    y = []
    for i in text.split(): 
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000, stop_words='english')
```

```python
new_df['tags'] 
vectors = cv.fit_transform(new_df['tags']).toarray()
new_df 
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
recommendation = []
def recommend(place):
    place_index = new_df[new_df['name']==place].index[0]
    distances = similarity[place_index]
    place_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]
    print(place_list)
    for i in place_list:
        recommendation.append(i)
    try:
        if not recommendation is not None:
            return recommendation 
    except Exception as e:
        return "recommendation not found"
```

```python
recommend('Phewa Lake')
```

```python
import pickle

pickle.dump(similarity, open('similarity.pkl','wb'))
pickle.dump(new_df, open('new_df.pkl','wb'))
```
